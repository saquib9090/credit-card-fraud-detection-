importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
import statistics
```

```python
# loading a dataset to a pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.tail()
```

```python
# datasets information
credit_card_data.info()
```

```python
# checking the missing values in the dataset in each column
credit_card_data.isnull().sum()
```

```python
# the distrubution of legit transactions and fraudulent transactions
credit_card_data['Class'].value_counts()
```

this dataset is highly unbalanced

0-----> normal Transaction
1-----> fraudulent Transaction


```python
# separting the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraudlent = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraudlent.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraudlent.Amount.describe()
```

```python
# compare the values both of these classes
credit_card_data.groupby('Class').mean()
```

under-sampling

build a sample dataset containing similar distribution of normal transactions and fraudlent transactions

the number of fraudlent transactions is 492 of 31 column

```python
legit_sample = legit.sample(n=492)
```

concatenating two dataframe

axis=0---->rows
axis=1---->column

```python
new_dataset = pd.concat([legit_sample,fraudlent],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and targets

```python
x = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

split the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

model training

use LogisticRegression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with Training data
model.fit(X_train, Y_train)
```

model Evaluation (perfomance of our model)

Accuracy score

```python
# accuracy on training data
X_train_prediction =  model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('accuracy on training data ;' , training_data_accuracy)
```

```python
X_test_prediction =  model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('accuracy on test data :' , test_data_accuracy)
```
